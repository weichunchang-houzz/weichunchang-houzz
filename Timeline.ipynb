```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_clipboard()
```

```python
data[:3]
```

```python
data.columns[3:].tolist()
```

```python
for col in data.columns[3:].tolist():

    temp_data = data.groupby(['time','ivy_company_type'])[col].sum().reset_index()
    sns.lineplot(data=temp_data, x='time',y=col,hue='ivy_company_type')
    plt.xticks(rotation=90)
    plt.title(f'{col} by company_type')
    plt.show()
```

```python
for col in data.columns[3:].tolist():

    temp_data = data.groupby(['time','country'])[col].sum().reset_index()
    sns.lineplot(data=temp_data, x='time',y=col,hue='country')
    plt.xticks(rotation=90)
    plt.title(f'{col} by country')
    plt.show()
```
